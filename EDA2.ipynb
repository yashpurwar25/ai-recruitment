```python
district_wealth = (
    df.groupby('District_Name')['Wealth_Index']
      .agg(['mean', 'median'])
      .sort_values('mean', ascending=False)
)

district_wealth
```

```python
style.use("dark_background")
district_wealth.plot(
    kind='bar',
    figsize=(10,5),
)

plt.ylabel('Wealth')
plt.title('Average and Median Wealth by District')
plt.xticks(rotation=45)
plt.show()
```

Q1-Compare district wise wealth index
Ans-The golden reef has the highest weakth while coral slums has the lowest wealth index
In all the districts we can see that kean wealth is higher than median wealth
the gap between mean and median shows the wealth inequality
Median income provides better comparative measure than mean income

```python
district_health = (
    df.groupby('District_Name')
      .agg(
          avg_wealth_index=('Wealth_Index', 'mean'),
          avg_life_expectancy=('Life_Expectancy', 'mean')
      )
)

district_health
```

```python
plt.figure(figsize=(7,5))
plt.plot(
    district_health['avg_wealth_index'],
    district_health['avg_life_expectancy'],c="red"
)

plt.xlabel('Average Wealth Index by District')
plt.ylabel('Average Life Expectancy by District')
plt.title('District Wealth Index vs Life Expectancy')
plt.show()
```

```python
occupation_income = (
    df.groupby('Occupation')['Wealth_Index']
      .agg(['mean', 'median'])
      .sort_values('median', ascending=False)
)

occupation_income
```

```python
occupation_income.plot(kind='bar',figsize=(10,5))
plt.ylabel('Income')
plt.title('Mean and Median Wealth Index by Occupation')
plt.xticks(rotation=45)
plt.show()
```
