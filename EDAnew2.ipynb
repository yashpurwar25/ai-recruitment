```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = df.drop(columns=["Occupation", "Citizen_ID"])
y = df["Occupation"]
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
X_train_enc = pd.get_dummies(X_train)
X_val_enc = pd.get_dummies(X_val)
# align columns
X_train_enc, X_val_enc = X_train_enc.align(X_val_enc, join="left", axis=1, fill_value=0)
```

```python
rf = RandomForestClassifier(n_estimators=10,random_state=20,class_weight="balanced")
rf.fit(X_train_enc, y_train)
```

```python
y_pred = rf.predict(X_val_enc)
cm = confusion_matrix(y_val, y_pred)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm,annot=True,fmt="d",cmap="Blues",xticklabels=sorted(y.unique()),yticklabels=sorted(y.unique()))
plt.xlabel("Predicted Occupation")
plt.ylabel("Actual Occupation")
plt.title("Confusion Matrix: Occupation Classification")
plt.show()
```

```python
cm_df = pd.DataFrame(cm, index=sorted(y.unique()), columns=sorted(y.unique()))
confused_pairs = (cm_df.where(~np.eye(cm_df.shape[0], dtype=bool)).stack().sort_values(ascending=False))
confused_pairs.head(5)
```

```python
df['Wealth_Bucket'] = pd.qcut(df['Wealth_Index'],q=3,labels=['Low', 'Medium', 'High'])
```

```python
age_summary = (df.groupby(['Wealth_Bucket', 'Occupation'])['Life_Expectancy'].mean().reset_index())
age_summary
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=age_summary,x='Occupation',y='Life_Expectancy',hue='Wealth_Bucket',marker='o')
plt.title("Interaction Between Age, Occupation, and Wealth Level")
plt.ylabel("Average Age")
plt.xlabel("Occupation")
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df,x='Occupation',y='Life_Expectancy',hue='Wealth_Bucket')
plt.title("Age Distribution Across Occupations by Wealth Level")
plt.ylabel("Age")
plt.xlabel("Occupation")
plt.show()
```

Q7-Does life expectancy interact with occupation differently for different wealth levels?
Ans-The analysis shows that life expectancy varies differently across occupations depending on wealth level.For higher wealth groups,certain occupations tend to have higher average ages,while in lower wealth groups the life expectancy distribution across occupations overlap more stronglyThis suggests an interaction effect between life expectancy,wealth and occupation,indicating that life expectancy alone is insufficient to distinguish occupations without considering wealth context. 

```python
# Create quantile-based bins
df['HouseSize_Bin'] = pd.qcut(df['House_Size_sq_ft'],q=4,labels=['Small', 'Medium', 'Large', 'Very Large'])
```

```python
house_occ_dist = pd.crosstab(df['HouseSize_Bin'],df['Occupation'],normalize='index')
house_occ_dist
```

```python
plt.figure(figsize=(9, 6))
sns.heatmap(house_occ_dist,annot=True,fmt=".2f",cmap="YlGnBu")
plt.title("Occupation Distribution Across House Size Ranges")
plt.xlabel("Occupation")
plt.ylabel("House Size Range")
plt.show()
```

```python
house_occ_dist.plot(kind='bar',stacked=True,figsize=(10, 6))
plt.title("Occupation Concentration by House Size Range")
plt.ylabel("Proportion")
plt.xlabel("House Size Range")
plt.legend(title="Occupation", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
